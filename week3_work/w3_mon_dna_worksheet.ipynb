#### Background

In molecular biology, **transcription** converts DNA to mRNA by swapping nucleotide bases:

| DNA Base     | mRNA Base    |
| ------------ | ------------ |
| A (Adenine)  | U (Uracil)   |
| T (Thymine)  | A (Adenine)  |
| G (Guanine)  | C (Cytosine) |
| C (Cytosine) | G (Guanine)  |

**Example:** DNA sequence `TCGTTCAGT` → mRNA sequence `AGCAAGUCA`
