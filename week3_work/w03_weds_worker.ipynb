## 1A. List of Lists 


````python
# Simple example

seasons = ["Spring", "Summer", "Fall", "Winter"]
fruits = ["Apple", "Banana", "Cherry"]

for season in seasons:
    print(season)
    print("========")
    for fruit in fruits:
        print(fruit, "in", season)
````
