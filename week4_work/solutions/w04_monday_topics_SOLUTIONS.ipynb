## Part 1: SOLUTION

```python
def create_temperature_converter(from_scale, to_scale):
    """Creates a custom temperature converter function"""
    
    def convert(value):
        """The actual conversion function"""
        # Convert based on the scales
        if from_scale == "C" and to_scale == "F":
            result = value * 9/5 + 32
        elif from_scale == "F" and to_scale == "C":
            result = (value - 32) * 5/9
        elif from_scale == "C" and to_scale == "K":
            result = value + 273.15
        elif from_scale == "K" and to_scale == "C":
            result = value - 273.15
        else:
            result = value  # Same scale
        
        return f"{value} {from_scale} = {result:.2f} {to_scale}"
    
    return convert

# Create specific converters
celsius_to_fahrenheit = create_temperature_converter("C", "F")
fahrenheit_to_celsius = create_temperature_converter("F", "C")

print(celsius_to_fahrenheit(5))  # 5 C = 41.00 F
print(fahrenheit_to_celsius(41))  # 41 F = 5.00 C
```

# Part 2: SOLUTION

```python
grades = [
    ["Alice", 85, 92, 88],
    ["Bob",   78, 85, 90],
    ["Carol", 92, 88, 95]
]

# Outer loop: iterate through rows (students)
for student in grades:
    name = student[0]
    scores = student[1:]  # All elements except name
    
    # Inner loop: iterate through columns (assignments)
    total = 0
    for score in scores:
        total += score
    
    average = total / len(scores)
    print(f"{name}'s average: {average:.1f}")
```

# Part 3: Dictionaries with Nested Functions

## Solution

```python
# Closure Example: Chinese Zodiac Calculator

def create_zodiac_calculator():
    """Create a calculator for Chinese zodiac years"""
    
    zodiacs = ["Rat", "Ox", "Tiger", "Rabbit", "Dragon", "Snake",
               "Horse", "Goat", "Monkey", "Rooster", "Dog", "Pig"]
    
    def year_to_zodiac(year):
        """Convert year to zodiac animal"""
        index = (year - 4) % 12  # 4 AD was a Rat year
        return zodiacs[index]
    
    def zodiac_to_years(zodiac, start_year, end_year):
        """Find all years in range for a zodiac"""
        years = []
        for year in range(start_year, end_year + 1):
            if year_to_zodiac(year) == zodiac:
                years.append(year)
        return years
    
    # Return a dictionary of operations
    return {
        "year_to_zodiac": year_to_zodiac,
        "zodiac_to_years": zodiac_to_years,
        "animals": zodiacs
    }

# Create the calculator
zodiac_calc = create_zodiac_calculator()

# Example usage:
print(zodiac_calc["year_to_zodiac"](2024))  # Find zodiac for 2024

# Need to find all years of the Dragon between 2000 and 2025
dragon_years = zodiac_calc["zodiac_to_years"]("Dragon", 2000, 2025)
print(f"Dragon years between 2000 and 2025: {dragon_years}")
```

# Part 4 Plotting SOLUTION

```python
import matplotlib.pyplot as plt

# Data for Blvand, Denmark (July 8-16)
dates = ['Jul 8', 'Jul 9', 'Jul 10', 'Jul 11', 'Jul 12', 
         'Jul 13', 'Jul 14', 'Jul 15', 'Jul 16']
air_temp = [19, 21, 22, 20, 18, 19, 21, 23, 22]      # Air temperature (C)
water_temp = [16, 16, 17, 17, 16, 16, 17, 17, 18]    # Water temperature (C)

# Create first subplot (Air Temperature)
plt.subplot(211)  # 2 rows, 1 column, position 1
plt.plot(dates, air_temp, 'o-', color='orangered', linewidth=2)
plt.ylabel('Air Temperature (C)')
plt.title('Blvand Temperature - July 8-16')
plt.grid(True, alpha=0.3)
plt.ylim(15, 25)

# Create second subplot (Water Temperature)
plt.subplot(212)  # 2 rows, 1 column, position 2
plt.plot(dates, water_temp, 's-', color='steelblue', linewidth=2)
plt.ylabel('Water Temperature (C)')
plt.xlabel('Date')
plt.grid(True, alpha=0.3)
plt.ylim(15, 25)

plt.tight_layout()  # Prevents overlapping labels
plt.show()
```
