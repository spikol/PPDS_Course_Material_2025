## Part 1: Nested Functions Examples

### Example 1: Simple Nested Function

```python
# Example 1: Simple Nested Function

def outer_function(name):
    """This is the outer function"""
    
    def inner_function():
        """This is nested inside outer_function"""
        return f"Hello from inside, {name}!"
    
    # Call the inner function
    result = inner_function()
    return result

# Call the outer function
print(outer_function("Alice"))
```

#### Example 2: Nested Function with Parameters

```python
# Example 2: Nested Function with Parameters

def process_temperature_data(readings):
    """Process temperature readings with internal helper"""
    
    def validate_reading(temp):
        """Helper function - only used inside this function"""
        return -50 <= temp <= 50
        
    
    valid_readings = [r for r in readings if validate_reading(r)]
    return sum(valid_readings) / len(valid_readings)

readings = [22, 25, -60, 30, 45, 55, 18]

average_temp = process_temperature_data(readings)
print(f"Average valid temperature: {average_temp}")
```

#### Example 3: Functions that remember

```python
# Example 3: Closure Example
def create_multiplier(factor):
    """Returns a function that multiplies by factor"""
    
    def multiply(number):
        # This inner function "remembers" factor
        return number * factor
    
    return multiply  # Return the function itself!

# Create specific multiplier functions
double = create_multiplier(2)
triple = create_multiplier(3)

print(double(5))   # 10
print(triple(5))   # 15
```

# Part 2: Nested Loops with Lists of Lists

#### Lists of Lists

```python
# Student grades - rows are students, columns are assignments
grades = [
    ["Alice", 85, 92, 88],    # Row 0
    ["Bob",   78, 85, 90],    # Row 1
    ["Carol", 92, 88, 95]     # Row 2
]
#  Column:  0     1   2   3

print(grades[0])        # ['Alice', 85, 92, 88] - first row
print(grades[1][2])     # 85 - Bob's second assignment
print(grades[0][1])     # 85 - Alice's first assignment
```

#### Real-World Example: Temperature Grid

```python
# Temperature readings from 4 sensors over 3 days
temp_data = [
    ["Mon", 22, 24, 23, 25],  # Day, Sensor1, Sensor2, Sensor3, Sensor4
    ["Tue", 21, 23, 22, 24],
    ["Wed", 23, 25, 24, 26]
]

print("Daily Averages:")
for day_reading in temp_data:
    day = day_reading[0]
    temps = day_reading[1:]
    avg = sum(temps) / len(temps)
    print(f"{day}: {avg:.1f}°C")

print("\nSensor Averages:")
for sensor_num in range(1, 5):
    temps = []
    for day_reading in temp_data:
        temps.append(day_reading[sensor_num])
    avg = sum(temps) / len(temps)
    print(f"Sensor {sensor_num}: {avg:.1f}°C")
```

```python
grades = [
    ["Alice", 85, 92, 88],
    ["Bob",   78, 85, 90],
    ["Carol", 92, 88, 95]
]

# Outer loop: iterate through rows (students)
for student in grades:
    name = student[_]
    scores = student[_:]  # All elements except name
    
    # Inner loop: iterate through columns (assignments)
    total = 0
    for score in scores:
        total += score
    
    average = total / ___(scores)
    print(f"{name}'s average: {average:.1f}")
```

# Part 3: Dictionaries with Nested Functions

#### Dictionaries can store functions as values!

```python
# Function definitions

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

# Store functions in a dictionary
calculator = {
    "add": add,
    "subtract": subtract,
    "multiply": multiply
}

# Use the dictionary to call functions
result = calculator["add"](5, 3)        # 8
result = calculator["multiply"](4, 2)   # 8
print(result)
```

#### Nested Functions Creating Dictionary Operations

```python
# Closure Example: Chinese Zodiac Calculator

def create_zodiac_calculator():
    """Create a calculator for Chinese zodiac years"""
    
    zodiacs = ["Rat", "Ox", "Tiger", "Rabbit", "Dragon", "Snake",
               "Horse", "Goat", "Monkey", "Rooster", "Dog", "Pig"]
    
    def year_to_zodiac(year):
        """Convert year to zodiac animal"""
        index = (year - 4) % 12  # 4 AD was a Rat year
        return zodiacs[index]
    
    def zodiac_to_years(zodiac, start_year, end_year):
        """Find all years in range for a zodiac"""
        years = []
        for year in range(start_year, end_year + 1):
            if year_to_zodiac(year) == zodiac:
                years.append(year)
        return years
    
    # Return a dictionary of operations
    return {
        "year_to_zodiac": year_to_zodiac,
        "zodiac_to_years": zodiac_to_years,
        "animals": zodiacs
    }

# Create the calculator
zodiac_calc = create_zodiac_calculator()

# Example usage:
print(zodiac_calc["year_to_zodiac"](2024))  # Find zodiac for 2024
```

#### Why Visualize Data?

```python
# Temperature data
temperatures = [22, 24, 23, 25, 24, 26, 27, 26, 25, 24]

# Numbers tell us something...
print(f"Average: {sum(temperatures)/len(temperatures)}")
print(f"Max: {max(temperatures)}")
print(f"Min: {min(temperatures)}")

# But a plot shows patterns instantly!
import matplotlib.pyplot as plt
plt.plot(temperatures)
plt.title("Daily Temperatures")
plt.xlabel("Day")
plt.ylabel("Temperature (°C)")
plt.show()
```

#### Bar Plots

```python
import matplotlib.pyplot as plt

# Data: Count of students born in each zodiac year (example data)
zodiacs = ["Rat", "Ox", "Tiger", "Rabbit", "Dragon", "Snake",
           "Horse", "Goat", "Monkey", "Rooster", "Dog", "Pig"]
counts = [3, 5, 2, 4, 6, 3, 4, 2, 5, 3, 4, 3]

# Create bar chart
plt.figure(figsize=(12, 6))
plt.bar(zodiacs, counts, color='steelblue')
plt.title('Class Distribution by Chinese Zodiac', fontsize=14)
plt.xlabel('Zodiac Animal', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)
plt.xticks(rotation=45)  # Rotate labels for readability

# Add value labels on bars
for i, v in enumerate(counts):
    plt.text(i, v + 0.1, str(v), ha='center')

plt.tight_layout()
plt.show()
```

#### Grid Layout & Subplot Positioning

```python
import matplotlib.pyplot as plt

# Data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
rainfall = [45, 38, 52, 48, 55, 62]
temperature = [2, 3, 7, 11, 15, 18]

# Create first subplot (top position)
plt.subplot(211)  # 2 rows, 1 column, position 1
plt.bar(months, rainfall, color='steelblue')
plt.ylabel('Rainfall (mm)')
plt.title('Copenhagen Weather Data')

# Create second subplot (bottom position)
plt.subplot(212)  # 2 rows, 1 column, position 2
plt.plot(months, temperature, 'o-', color='orangered', linewidth=2)
plt.ylabel('Temperature (°C)')
plt.xlabel('Month')

plt.tight_layout()  # Prevents overlapping labels
plt.show()
```

#### Multi-line Plots

```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales_2023 = [100, 150, 130, 170, 160]
sales_2024 = [120, 160, 150, 190, 180]

# Plot both lines
plt.plot(months, sales_2023, marker='o', label='2023')
plt.plot(months, sales_2024, marker='s', label='2024')

plt.title('Sales Comparison')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.legend()
plt.grid(True, alpha=0.3)

plt.show()
```
